# World Wide Products Inc. - Forecasting Product Demand

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima_model import ARIMA

from hmmlearn.hmm import GaussianHMM
```

```python
data = pd.read_csv('../data/Historical Product Demand.csv')
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

## Preprocessing

#### Slicing the target product demand

```python
# Product Details for which we wish to forecast deman

# product_code = 'Product_0001'
# warehouse = 'Whse_A'
# product_category = 'Category_005'

product_code = 'Product_0979'
warehouse = 'Whse_J'
product_category = 'Category_028'

product_data = data[(data['Product_Code']==product_code) & (data['Warehouse']==warehouse) & (data['Product_Category']==product_category)][['Date', 'Order_Demand']]
product_data.index = product_data['Date']
product_data.drop(['Date'], inplace=True, axis=1)

product_data.head()
```

#### Cleaning Data

1. Removing non-numerical characters
2. Replacing infinite values by the mean
3. Sorting by date

```python
product_data['Order_Demand'].replace('[^0-9]', '', regex=True, inplace=True)
product_data['Order_Demand'] = product_data['Order_Demand'].astype(np.int64)
```

```python
boolean_array = product_data[(np.isinf(product_data['Order_Demand']))]
mean_array = np.full((len(boolean_array), 1), product_data[~np.isinf(product_data['Order_Demand'])].mean())

product_data[(np.isinf(product_data['Order_Demand']))] = mean_array
```

```python
product_data.sort_values(by=['Date'], inplace=True)
```

#### Mean Smoothing by the Months

```python
product_data_mean_by_month = product_data.groupby([product_data.index.year, product_data.index.month]).mean()

product_data_index = [pd.to_datetime('{}-{}-01'.format(i, j)) for i, j in product_data_mean_by_month.index]
product_data_mean_by_month.index = product_data_index

product_data = product_data_mean_by_month
```

#### Standardizing the data

```python
initial_avg, initial_dev = product_data['Order_Demand'].mean(), product_data['Order_Demand'].std()

product_data['Order_Demand'] = (product_data['Order_Demand']-initial_avg)/initial_dev
```

#### Taking the first differences to improve stationarity

```python
product_data['Order_Demand'] = product_data['Order_Demand'].diff()
product_data.dropna(inplace=True)
```

```python
fig = plt.gcf()
fig.set_size_inches(18, 5)

plt.xlabel('Date')
plt.ylabel('Order_Demand')
plt.title('Product: ' + product_code + ' - Warehouse: ' + warehouse + ' - Product Category: ' + product_category)
plt.plot(product_data)

plt.axhline(product_data['Order_Demand'].mean(), linestyle='--', color='k', alpha=0.3)

for d in list(map(lambda x: str(x)+'-01-01', [y for y in range(2012, 2018)])):
    plt.axvline(pd.to_datetime(d), linestyle='--', color='k', alpha=0.3)
```

### Testing Stationarity of the time series data

```python
def rolling_statistics(data):
    rolmean = data['Order_Demand'].rolling(window=12).mean()
    rolstd = data['Order_Demand'].rolling(window=12).std()
    
    return rolmean, rolstd
    
def adf_test(data):
    dfTest = adfuller(data['Order_Demand'], autolag='AIC')
    
    dfOutput = pd.DataFrame([dfTest[0:4]], columns=['test-statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dfTest[4].items():
        dfOutput['Critical Value (%s)'%key] = value
    
    return dfOutput
    
def test_stationarity(data):
    # Rolling mean test
    mean, std = rolling_statistics(data)
    
    fig = plt.gcf()
    fig.set_size_inches(18, 10)
    
    plt.xlabel('Date')
    plt.ylabel('Order_Demand')
    
    orig = plt.plot(data, color='blue', label='Original')
    mean = plt.plot(mean, color='red', label='Rolling Mean')
    std = plt.plot(std, color='black', label='Rolling Std')
    
    plt.legend(loc='best')
    
    plt.axhline(product_data['Order_Demand'].mean(), linestyle='--', color='k', alpha=0.3)

    for d in list(map(lambda x: str(x)+'-01-01', [y for y in range(2012, 2018)])):
        plt.axvline(pd.to_datetime(d), linestyle='--', color='k', alpha=0.3)
    
    # Augmented Dickey-Fuller Test
    adf_output = adf_test(data)
    
    print(adf_output.head())
    
    # Test Stationarity
    val1 = (adf_output['test-statistic'] < adf_output['Critical Value (1%)'])
    val2 = (adf_output['test-statistic'] < adf_output['Critical Value (5%)'])
    val3 = (adf_output['test-statistic'] < adf_output['Critical Value (10%)'])
    
    return (val1 & val2 & val3).squeeze()
```

```python
if(test_stationarity(product_data)):
    print('Product Data is Stationary')
else:
    print('Product Data is not Stationary')
```

#### Log Transformation for improving stationarity

This step is performed to improve stationarity by tempering down variance and mean

```python
product_data['Order_Demand'] = np.sign(product_data['Order_Demand']) * np.log(abs(product_data['Order_Demand']) + 1)
product_data.dropna(inplace=True)

test_stationarity(product_data)
```

#### Standardizing log transformed data for improving stationarity

This step is performed to imporove stationarity by tempering down variance and mean within each year's timeframe.

```python
annual_mean = product_data.groupby(product_data.index.year).mean()
annual_volatility = product_data.groupby(product_data.index.year).std()

product_mean = product_data.index.map(lambda d: annual_mean.loc[d.year].squeeze()).to_numpy().reshape(-1, 1)
product_std = product_data.index.map(lambda d: annual_volatility.loc[d.year].squeeze()).to_numpy().reshape(-1, 1)

product_data = (product_data-product_mean)/product_std
```

```python
print('\nData Stationarity: ', test_stationarity(product_data), '\n')
```

#### Exploring trend, seasonality and the residual of the decomposed data

```python
decomposition = seasonal_decompose(product_data, period=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.gcf().set_size_inches(18, 8)

plt.subplot(4, 1, 1)
plt.plot(product_data, label='Original (normalized, log)')
plt.legend(loc='best')

plt.subplot(4, 1, 2)
plt.plot(trend, label='Trend')
plt.legend(loc='best')

plt.subplot(4, 1, 3)
plt.plot(seasonal, label='Seasonality')
plt.legend(loc='best')

plt.subplot(4, 1, 4)
plt.plot(residual, label='Residual')
plt.legend(loc='best')
```

## Forecasting

```python
# This forecast will be on a scale according to the sequence of transformations done for preprocessing
# This function will invert the scaled forecast to the original order demand unit
# This function will be called at the end of both ARIMA and HMM forecasts

def invert_forecast_to_original_scale(forecast):
    annual_mean = product_data.groupby(product_data.index.year).mean()
    annual_volatility = product_data.groupby(product_data.index.year).std()

    product_mean = product_data.index.map(lambda d: annual_mean.loc[d.year].squeeze()).to_numpy().reshape(-1, 1)
    product_std = product_data.index.map(lambda d: annual_volatility.loc[d.year].squeeze()).to_numpy().reshape(-1, 1)

    std = product_std[-1].squeeze()
    mean = product_mean[-1].squeeze()
    
    forecast = (forecast * std) + mean
    
    forecast = np.exp(forecast) - 1
    
    forecast = (forecast * initial_dev) + initial_avg
    
    return forecast
```

### ARIMA

```python
lag_acf = acf(product_data)
lag_pacf = pacf(product_data, method='ols')

plt.gcf().set_size_inches(18, 8)

plt.subplot(2, 1, 1)
plt.plot(lag_acf)

plt.axhline(0, linestyle='--', color='gray')
plt.title('Autocorrelation Function\n(Q-value for ARIMA - at correlation value 0)')

plt.subplot(2, 1, 2)
plt.plot(lag_pacf)

plt.axhline(0, linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function\n(P-value for ARIMA - at correlation value 0)')

plt.tight_layout()
```

```python
# P value from PACF plot
p_value = 0

# Q value from ACF plot
q_value = 1

# Number of times series difference is taken
d_value = 0

model_arima = ARIMA(product_data, order=(p_value, d_value, q_value))
result = model_arima.fit(disp=-1)

plt.gcf().set_size_inches(18, 8)

plt.plot(product_data, label='Original')
plt.plot(result.fittedvalues, color='red', label='ARIMA Generated')

plt.legend(loc='best')

sse = sum((result.fittedvalues-product_data['Order_Demand'])**2)
mse = sse/len(result.fittedvalues)

plt.title('Mean Squared Error for ARIMA order (0, 1, 0): %.4f'% mse)
```

```python
forecast_ARIMA = result.forecast(steps=1)

print('ARIMA forecast: {}'.format(invert_forecast_to_original_scale(forecast_ARIMA[0]).squeeze()))
```

### Hidden Markov Model

```python
num_components = 4

model_hmm = GaussianHMM(n_components = num_components, covariance_type = 'full', n_iter = 10000, algorithm='viterbi')

model_hmm.fit(np.expand_dims(product_data['Order_Demand'].to_numpy(), axis=1))
```

```python
hmm_predicted = pd.DataFrame(model_hmm.predict(product_data), columns=['Forecast'], index=product_data.index)

# Standardizing the output
hmm_predicted = (hmm_predicted - hmm_predicted.mean())/hmm_predicted.std()

plt.gcf().set_size_inches(18, 8)

plt.plot(product_data, label='Original')
plt.plot(hmm_predicted, color='red', label='HMM Generated')

plt.legend(loc='best')

sse = sum((hmm_predicted['Forecast']-product_data['Order_Demand'])**2)
mse = sse/len(hmm_predicted)

plt.title('Mean Squared Error for HMM with 4 components: %.4f'% mse)
```

```python
#create DatetimeIndex by last value of index
new_idx = pd.to_datetime(product_data.index[-1] + pd.Timedelta(31, 'days'))

for i in range(model_hmm.n_components):
    
    # g(x) = (1/(sigma * sqrt(2*pi))) * e^((-(x-mu)^2/2*sigma^2))
    # Taking the gaussian sample with the highest probability at the mean
    
    forecast = 1/(model_hmm.covars_[i].squeeze().squeeze() * np.sqrt(2 * np.pi))
    
    print('For the next likely state as the hidden state %d, forecasted value: %.4f'%(i, invert_forecast_to_original_scale(forecast)))

    new_forecast = pd.DataFrame([forecast], index=[new_idx])
    new_forecast = hmm_predicted.append(new_forecast)
    
    plt.gcf().set_size_inches(18, 8)
    
    ax = plt.subplot(model_hmm.n_components, 1, i+1)
    ax.set_title('Using hidden state {} as the next likely state'.format(i))
    
    plt.plot(product_data, label='Original')
    plt.plot(new_forecast, color='red', label='HMM Generated')

    plt.legend(loc='lower left')
    plt.tight_layout()
```
